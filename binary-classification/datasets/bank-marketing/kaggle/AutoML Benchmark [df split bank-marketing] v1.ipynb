```python
NAME = 'bank-marketing'
```

```python
import sklearn.model_selection
```

```python
data= pd.read_csv(f'/kaggle/input/automl-benchmark/binary-classification/data/{NAME}/{NAME}.csv')
data
```

```python
data["Class"] = data["Class"].astype('category').cat.codes  
```

```python
data
```

```python
X = data.drop('Class',axis=1)
Y = data['Class']
```

```python
X
```

```python
X_train, X_test, y_train, y_test = sklearn.model_selection.train_test_split(X, Y, 
                                             test_size=0.25, 
                                             random_state=42)
X_train.shape, X_test.shape
```

```python
X_test
```

```python
X_train = X_train.reset_index(drop=True)
X_test = X_test.reset_index(drop=True)
```

```python
X_train['class'] = y_train.reset_index(drop=True)
```

```python
y_test_ = X_test.copy()
y_test_['class'] = y_test.reset_index(drop=True)
y_test_['id'] = X_test.index
y_test = y_test_[['id', 'class']]
```

```python
X_train.to_csv('train.csv', index=False,)
X_test.to_csv('test.csv', index=False,)
y_test.to_csv('solution.csv', index=False,)
```
